Criar um sistema bancário com as operações de sacar, depositar e visualizar o extrato.

Deposito:
- Deve ser possível depositar valores positivos para a minha conta bancária
- Todos depósitos devem ser armazenados em uma variável e exibidos na operação de extrato.

Sacar:
- Deve permitir realizar 3 saques diários com limite máximo de R$ 500,00 por saque.
- Caso o usuário não tenha saldo em conta, o sistema deve exibir uma mensagem informando que não será possível sacar o dinheiro por falta de saldo
- Todos os saques devem ser armazenados em uma variável e exibidos no extrato.

Extrato:
- Deve listar todos os depósitos e saques realizados em conta.
- No fim da listagem deve ser exibido o saldo atual da conta
- Os valores devem ser exibidos utilizando o formato R$ XXX.XX


```python
def depositar(valor):
    global saldo
    if 0 < valor:
        registros_depositos.append(valor)
        saldo += valor
        print(f"O seu depósito de R${valor: ,.2f} foi realizado com sucesso. \n\n"
              f"Seu saldo agora é de R${saldo: ,.2f}")
    else:
        print("O valor do seu depósito deve ser maior que 0.")
        
    return
        
def sacar(valor):  
    global saldo, saque
    if valor <= 500: 
        saque += 1
        print(f"Esse é o seu saque de número: {saque}")
        if saque <= 3:
            # print("Qual é o valor que gostaria de sacar?")
            novo_saldo = saldo - valor
            if 0 <= novo_saldo:
                registros_saques.append(valor)
                saldo = novo_saldo
                print(f"O seu saque de R${valor: ,.2f} foi realizado com sucesso. \n")
                print(f"Seu saldo agora é de R${saldo: ,.2f}")
            else:
                print("Você não pode realizar esse valor de saque, "
                      f"pois seu saldo é de R${saldo: ,.2f}")
        else:
            print("Você não pode realizar mais saques, pois atingiu o limite de 3 saques diários.")
    else:
        print("Você não pode realizar um saque de valor maior que R$ 500,00.")
        
def retirar_extrato():
    global saldo
    
    print("Você realizou os seguintes depósitos:\n")
    for deposito in registros_depositos:
        print(f"R${deposito: ,.2f} \n")
        
    print("Você realizou os seguintes saques:\n")
    for saque in registros_saques:
        print(f"R${saque: ,.2f} \n")
    
    print(f"Seu saldo atual é de R${saldo: ,.2f}")


    
menu = """
Seja bem vindo ao Banco. Escolha uma das ações a seguir para continuar:

[d] Depositar
[s] Sacar
[e] Extrato
[q] Sair

=> """

saldo = 0
limite_saque = 500
extrato = ""
NUMEROS_SAQUE = 3
saque = 0

registros_depositos = []
registros_saques = []

while True:
    
    opcao = input(menu)
    
    if opcao == "d":
        print("Qual é o valor que gostaria de depositar?")
        valor = input("Depósito: ")
        depositar(float(valor))
        
    elif opcao == "s":
        valor = print("Qual é o valor que gostaria de sacar? ")   
        valor = input("Saque: ")
        sacar(float(valor))
        
    elif opcao == "e":
        print("------ EXTRATO ------ \n")
        retirar_extrato()
        
    elif opcao == "q":
        break

    else:
        print("Operação invalida, selecione uma operação válida")         
```
